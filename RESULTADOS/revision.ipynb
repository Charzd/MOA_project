## Punto de referencia

```python
#Calculo del punto de referencia
def calcular_punto_referencia(archivos):
    max_f1 = -float("inf")
    max_f2 = -float("inf")

    for archivo in archivos:
        with open(archivo, "r") as f:
            for linea in f:
                if linea.strip() == "":
                    continue
                try:
                    f1, f2 = map(float, linea.strip().split())
                    max_f1 = max(max_f1, f1)
                    max_f2 = max(max_f2, f2)
                except ValueError:
                    continue 
    
    ref_f1 = max_f1 * 1.10  
    ref_f2 = max_f2 * 1.10
    return [ref_f1, ref_f2]


#Pruebas de uso
archivos_caso1 = [
    "AMPL/ampl_caso1.txt",
    "NSGA/CASO1/sem1.txt", "NSGA/CASO1/sem2.txt", "NSGA/CASO1/sem3.txt",
    "NSGA/CASO1/sem4.txt", "NSGA/CASO1/sem5.txt", "NSGA/CASO1/sem6.txt",
    "NSGA/CASO1/sem7.txt", "NSGA/CASO1/sem8.txt", "NSGA/CASO1/sem9.txt",
    "NSGA/CASO1/sem10.txt"
]
archivos_caso2 = [
    "AMPL/ampl_caso2.txt",
    "NSGA/CASO2/sem1.txt", "NSGA/CASO2/sem2.txt", "NSGA/CASO2/sem3.txt",
    "NSGA/CASO2/sem4.txt", "NSGA/CASO2/sem5.txt", "NSGA/CASO2/sem6.txt",
    "NSGA/CASO2/sem7.txt", "NSGA/CASO2/sem8.txt", "NSGA/CASO2/sem9.txt",
    "NSGA/CASO2/sem10.txt"
]
ref_point_caso1 = calcular_punto_referencia(archivos_caso1)
print("Punto de referencia caso 1 (max +10%):", ref_point_caso1)
ref_point_caso2 = calcular_punto_referencia(archivos_caso2)
print("Punto de referencia caso 2 (max +10%):", ref_point_caso2)

```

## Normalizaci√≥n de los casos y funci√≥n de hipervolumen

```python
def normalizar_frente_caso1(frente, max1=44850, max2=15910400):
    frente_normalizado = [(x / max1, y / max2) for x, y in frente]
    return frente_normalizado

def normalizar_frente_caso2(frente, max1=35225.6, max2=28108800):
    frente_normalizado = [(x / max1, y / max2) for x, y in frente]
    return frente_normalizado

def calc_hipervolumen(frente,ref_point):
    frente_ordenado = sorted(frente, key=lambda x: x[0])
    
    hv = 0
    i = 0
    for i in range(len(frente_ordenado)):
        x, y = frente_ordenado[i]
        if i == 0:
            ancho = ref_point[0] - x
            alto  = ref_point[1] - y
        else:
            ancho = ref_point[0] - x
            alto = frente_ordenado[i-1][1] - y
        hv += ancho*alto
        
    return hv
```

## Hipervol√∫menes

```python
#Se puede cambiar para normalizar o no normalizar los frentes

files = ["AMPL/ampl_caso1.txt","AMPL/ampl_caso2.txt",
         "NSGA/CASO1/sem1.txt","NSGA/CASO1/sem2.txt","NSGA/CASO1/sem3.txt","NSGA/CASO1/sem4.txt","NSGA/CASO1/sem5.txt",
         "NSGA/CASO1/sem6.txt","NSGA/CASO1/sem7.txt","NSGA/CASO1/sem8.txt","NSGA/CASO1/sem9.txt","NSGA/CASO1/sem10.txt",
         "NSGA/CASO2/sem1.txt","NSGA/CASO2/sem2.txt","NSGA/CASO2/sem3.txt","NSGA/CASO2/sem4.txt","NSGA/CASO2/sem5.txt",
         "NSGA/CASO2/sem6.txt","NSGA/CASO2/sem7.txt","NSGA/CASO2/sem8.txt","NSGA/CASO2/sem9.txt","NSGA/CASO2/sem10.txt",]


for filename in files:
    if(filename in archivos_caso1):
        ref_point = calcular_punto_referencia(archivos_caso1)
    if(filename in archivos_caso2):
        ref_point = calcular_punto_referencia(archivos_caso2)
        
    frente = []
    file = open(filename, "r")
    linea = file.readline().strip().split()
    t = linea[2]
    
    
    for lineas in file:
        linea = lineas.strip().split()
        v1, v2 = linea
        frente.append((float(v1), float(v2)))
    file.close()
    
    #Si se quieren los frentes normalizados, se cambia el valor de abajo
    normalizar = False
    
    if normalizar:
        ref_point = (1.1, 1.1)
        if(filename in archivos_caso1):
            frente = normalizar_frente_caso1(frente)
        if(filename in archivos_caso2):
            frente = normalizar_frente_caso2(frente)
    
    
    hv = calc_hipervolumen(frente,ref_point)
    if filename[-5] == '0':
        print("Frende de: ", filename, "Tiempo: ", t, "Hipervolumen: ", hv)    
    else:
        print("Frende de: ", filename, " Tiempo: ", t, "Hipervolumen: ", hv)
    if (filename[0] == 'A' and filename[-5] == '2') or filename[-5]=='0':
         print()
```
